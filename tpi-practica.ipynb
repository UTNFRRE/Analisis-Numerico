3.b

Determine la solución general del sistema de ecuaciones (Autovalores distintos)



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Matriz del sistema
A = np.array([[2, 2],
              [1, 3]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
for i in eigvals:
    print(i)
print("Autovectores:")
for i in eigvecs:
    print(i)
# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = 2*x + 2*y
    dydt = x + 3*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [0, 0]

# Intervalo de tiempo
t = np.linspace(0, 10, 400)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = 2*X + 2*Y
V = X + 3*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Resuelva los siguientes sistemas no homogéneos por el método de variación de parámetros





```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la matriz A y el término no homogéneo
A = np.array([[1, 8], [1, -1]])
def non_homogeneous_term(t):
    return np.array([np.exp(-t), t * np.exp(t)])

# Paso 1: Resolver el sistema homogéneo
eigvals, eigvecs = la.eig(A)

print("Autovalores:")
for i in eigvals:
    print(i)
print("Autovectores:")
for i in eigvecs:
    print(i)

# Matriz de solución general del sistema homogéneo
def homogeneous_solution(t, c1, c2):
    return c1 * eigvecs[:, 0] * np.exp(eigvals[0] * t) + c2 * eigvecs[:, 1] * np.exp(eigvals[1] * t)

# Paso 2: Resolver el sistema no homogéneo por variación de parámetros
def system(t, X):
    return A @ X + non_homogeneous_term(t)

# Condiciones iniciales
X0 = [0, 0]

# Resolver usando scipy.integrate.solve_ivp
t_span = [0, 10]
t_eval = np.linspace(t_span[0], t_span[1], 200)

sol = solve_ivp(system, t_span, X0, t_eval=t_eval)

# Graficar las soluciones
plt.plot(sol.t, sol.y[0], label='x1(t)', color='blue')
plt.plot(sol.t, sol.y[1], label='x2(t)', color='red')
plt.title("Solución del sistema no homogéneo")
plt.xlabel("Tiempo t")
plt.ylabel("Funciones X(t)")
plt.legend()
plt.grid()
plt.show()
```

8.b Resolver los siguientes sistemas de ecuaciones con valores iniciales





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir la matriz A y las funciones no homogéneas f(t) y g(t)
A = np.array([[1, -1], [2, 4]])
def f(t):
    return np.exp(2 * t)

def g(t):
    return 1

# Definir el sistema de ecuaciones diferenciales
def system(t, X):
    x, y = X
    return A @ np.array([x, y]) + np.array([f(t), g(t)])

# Condiciones iniciales
X0 = [1, 2]  # x(0) = 1, y(0) = 2

# Intervalo de tiempo y puntos de evaluación
t_span = [0, 5]  # Queremos la solución entre t=0 y t=5
t_eval = np.linspace(t_span[0], t_span[1], 200)  # Evaluar en 200 puntos

# Resolver el sistema
sol = solve_ivp(system, t_span, X0, t_eval=t_eval)

# Graficar las soluciones
plt.plot(sol.t, sol.y[0], label='x(t)', color='blue')
plt.plot(sol.t, sol.y[1], label='y(t)', color='red')
plt.title("Solución del sistema no homogéneo")
plt.xlabel("Tiempo t")
plt.ylabel("Funciones x(t), y(t)")
plt.legend()
plt.grid()
plt.show()
```
