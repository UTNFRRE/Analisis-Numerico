3.b

Determine la solución general del sistema de ecuaciones (Autovalores distintos)



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Matriz del sistema
A = np.array([[2, 2],
              [1, 3]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
print(eigvals)
print("Autovectores:")
print(eigvecs)

# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = 2*x + 2*y
    dydt = x + 3*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [1, 1]

# Intervalo de tiempo
t = np.linspace(0, 10, 400)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = 2*X + 2*Y
V = X + 3*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Determine la solución general del sistema de ecuaciones (Autovalores repetidos)




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Matriz del sistema
A = np.array([[-6, 5],
              [-5, 4]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
print(eigvals)
print("Autovectores:")
print(eigvecs)

# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = -6*x + 5*y
    dydt = -5*x + 4*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [1, 1]

# Intervalo de tiempo
t = np.linspace(0, 10, 400)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = -6*X + 5*Y
V = -5*X + 4*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
