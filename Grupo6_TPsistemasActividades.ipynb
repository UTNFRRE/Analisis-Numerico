3.b

Determine la solución general del sistema de ecuaciones (Autovalores distintos)



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sympy import symbols, exp

# Matriz del sistema
A = np.array([[2, 2],
              [1, 3]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
for i in eigvals:
    print(i)
print("Autovectores:")
for i in eigvecs:
    print(i)

# Planteo simbólico de la solución general
t = symbols('t')
C1, C2 = symbols('C1 C2')
sol_general = C1 * (eigvecs[:, 0] * exp(eigvals[0] * t)) + \
              C2 * (eigvecs[:, 1] * exp(eigvals[1] * t))

print("\nSolución general:")
print(sol_general)


# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = 2*x + 2*y
    dydt = x + 3*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [0, 0]

# Intervalo de tiempo
t = np.linspace(0, 10, 200)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = 2*X + 2*Y
V = X + 3*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Como  𝜆1 > 0 y 𝜆2 > 0 entonces  el sistema es **inestable** y las soluciones se alejan del punto de equilibrio.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sympy import symbols, exp, I, cos, sin, Matrix, simplify

# Matriz del sistema
A = np.array([[1, 1],
              [-2, -1]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
for i in eigvals:
    print(f"{i.real:.2f} + {i.imag:.2f}j")
print("Autovectores:")
for i in eigvecs:
    print(i)

t = symbols('t')
C1, C2 = symbols('C1 C2')
alpha = np.real(eigvals[0])
beta = np.imag(eigvals[0])
v_real = np.real(eigvecs[:, 0])

# Solución general usando la forma exponencial y trigonométrica
sol_general = exp(alpha * t) * (C1 * cos(beta * t) + C2 * sin(beta * t)) * Matrix(v_real)
sol_general = simplify(sol_general)

print("\nSolución general:")
print(sol_general)

# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = -x + y
    dydt = -2*x + -1*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [0, 0]

# Intervalo de tiempo
t = np.linspace(0, 10, 400)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = -X + Y
V = -2*X + -1*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sympy import symbols, exp, Matrix, solve

# Matriz del sistema
A = np.array([[6, -1],
              [5, 4]])

# Autovalores y autovectores
eigvals, eigvecs = np.linalg.eig(A)

print("Autovalores:")
for i in eigvals:
    print(f"{i.real:.2f} + {i.imag:.2f}j")
print("Autovectores:")
for i in eigvecs:
    print(i)


# Planteo simbólico de la solución general y particular
# Definimos las variables simbólicas y la solución general
t = symbols('t')
C1, C2 = symbols('C1 C2')
x0, y0 = -2, 8  # Valores iniciales

# Convert eigenvectors to sympy matrices
eigvecs_sympy = Matrix(eigvecs)

# Solución general en términos de los vectores propios y valores propios
sol_general = C1 * eigvecs_sympy[:, 0] * exp(eigvals[0] * t) + \
              C2 * eigvecs_sympy[:, 1] * exp(eigvals[1] * t)

# Planteamos el sistema de ecuaciones para las condiciones iniciales
eqs = [
    sol_general[0].subs(t, 0) - x0,
    sol_general[1].subs(t, 0) - y0
]

# Resolvemos para C1 y C2
constants = solve(eqs, (C1, C2))

# Solución particular
sol_particular = sol_general.subs(constants)

print("\nSolución particular:")
print(sol_particular)

# Definir el sistema de ecuaciones diferenciales
def sistema(X, t):
    x, y = X
    dxdt = 6*x + -1*y
    dydt = 5*x + 4*y
    return [dxdt, dydt]

# Valores iniciales
X0 = [-2, 8]

# Intervalo de tiempo
t = np.linspace(0, 10, 400)

# Resolver el sistema de ecuaciones diferenciales
sol = odeint(sistema, X0, t)

# Graficar las soluciones
plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.title('Solución de x(t) y y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.legend()
plt.grid(True)
plt.show()

# Campo vectorial
Y, X = np.mgrid[-10:10:20j, -10:10:20j]
U = 6*X + -1*Y
V = 5*X + 4*Y

plt.quiver(X, Y, U, V, color='r')
plt.title('Campo vectorial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Como 𝜆1 = 𝛼 + 𝛽𝑖 𝜆2 = 𝛼 − 𝛽𝑖 el punto crítico se llama un **punto espiral y es inestable** ya que 𝛼 > 0

Resuelva los siguientes sistemas no homogéneos por el método de variación de parámetros





```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la matriz A y el término no homogéneo
A = np.array([[1, 8], [1, -1]])
def non_homogeneous_term(t):
    return np.array([np.exp(-t), t * np.exp(t)])

# Paso 1: Resolver el sistema homogéneo
eigvals, eigvecs = la.eig(A)

print("Autovalores:")
for i in eigvals:
    print(i)
print("Autovectores:")
for i in eigvecs:
    print(i)

# Matriz de solución general del sistema homogéneo
def homogeneous_solution(t, c1, c2):
    return c1 * eigvecs[:, 0] * np.exp(eigvals[0] * t) + c2 * eigvecs[:, 1] * np.exp(eigvals[1] * t)

# Paso 2: Resolver el sistema no homogéneo por variación de parámetros
def system(t, X):
    return A @ X + non_homogeneous_term(t)

# Condiciones iniciales
X0 = [0, 0]

# Resolver usando scipy.integrate.solve_ivp
t_span = [0, 10]
t_eval = np.linspace(t_span[0], t_span[1], 200)

sol = solve_ivp(system, t_span, X0, t_eval=t_eval)

# Graficar las soluciones
plt.plot(sol.t, sol.y[0], label='x1(t)', color='blue')
plt.plot(sol.t, sol.y[1], label='x2(t)', color='red')
plt.title("Solución del sistema no homogéneo")
plt.xlabel("Tiempo t")
plt.ylabel("Funciones X(t)")
plt.legend()
plt.grid()
plt.show()
```

8.b Resolver los siguientes sistemas de ecuaciones con valores iniciales





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir la matriz A y las funciones no homogéneas f(t) y g(t)
A = np.array([[1, -1], [2, 4]])
def f(t):
    return np.exp(2 * t)

def g(t):
    return 1

# Definir el sistema de ecuaciones diferenciales
def system(t, X):
    x, y = X
    return A @ np.array([x, y]) + np.array([f(t), g(t)])

# Condiciones iniciales
X0 = [1, 2]  # x(0) = 1, y(0) = 2

# Intervalo de tiempo y puntos de evaluación
t_span = [0, 5]  # Queremos la solución entre t=0 y t=5
t_eval = np.linspace(t_span[0], t_span[1], 200)  # Evaluar en 200 puntos

# Resolver el sistema
sol = solve_ivp(system, t_span, X0, t_eval=t_eval)

# Graficar las soluciones
plt.plot(sol.t, sol.y[0], label='x(t)', color='blue')
plt.plot(sol.t, sol.y[1], label='y(t)', color='red')
plt.title("Solución del sistema no homogéneo")
plt.xlabel("Tiempo t")
plt.ylabel("Funciones x(t), y(t)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.linalg import eig

# Define the function to analyze stability
def analyze_stability(k_values):
    for k in k_values:
        # Define matrix A for each value of k
        A = np.array([[0, 1], [-1, -k]])
        
        # Compute eigenvalues of A
        eigenvalues, _ = eig(A)
        
        # Analyze the real part of the eigenvalues
        stability = all(e.real < 0 for e in eigenvalues)
        
        # Output results
        print(f"k = {k}: autovalores = {eigenvalues}, estabilidad: {stability}")

# List of k values to analyze
k_values = [3, 2, 1, 0, -1, -2, -3]  # Examples covering all specified cases

# Run stability analysis
analyze_stability(k_values)
```
